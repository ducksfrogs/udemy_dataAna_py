```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(5,100,100)
y = np.linspace(10,1000,100)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x, y)

plt.xlabel("looko")
plt.ylabel("looko")
plt.title("rens")

plt.xlim([20,80])
plt.ylim([200,800])
```

```python
x = np.linspace(0,10,20)
y = x*2

plt.plot(x, y, 'b+')

plt.xlabel('Current')
plt.ylabel('Voltage')
plt.title("Ohm's law")
```

```python
x = np.linspace(0,5,10)
y = np.linspace(3,6,10)

plt.plot(x,y, 'r-', x,y**2, 'b+', x,y**3,'g^')
```

```python
x = np.linspace(1,10,100)
y = np.log(x)

plt.figure(1)

plt.subplot(121)
plt.title("y=log(x)")
plt.plot(x,y)

plt.subplot(122)
plt.title("y=log(x)**@")
plt.plot(x, y**2)


plt.show()
```

```python
x = np.linspace(1,10,100)

plt.figure(1)

plt.subplot(2,2,1)
plt.title("Linear")
plt.plot(x, x)

plt.subplot(2,2,2)
plt.title("Cubic")
plt.plot(x, x**3)

plt.figure(2)
plt.subplot(2,2,1)
plt.title("Log")
plt.plot(x, np.log(x))
plt.subplot(2,2,2)
plt.title("sqrt")
plt.plot(x, np.sqrt(x))
```
