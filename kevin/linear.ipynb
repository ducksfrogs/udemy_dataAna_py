```python
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchseAmount = 100- (pageSpeeds + np.random.normal(0, 0.1, 1000))*3

plt.scatter(pageSpeeds, purchseAmount)
```

```python
from scipy import stats
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchseAmount)
```

```python
r_value**2
```

```python
stats.corr(pageSpeeds, purchseAmount)
```

```python
def predict(x):
    return slope *x + intercept

fitLine = predict(pageSpeeds)
plt.scatter(pageSpeeds, purchseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```
